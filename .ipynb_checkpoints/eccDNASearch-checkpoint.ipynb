```python
# search for boundary using bigwig and macs bed file
import os, sys
import pysam
from pysam import AlignedSegment


complement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', '-':'-'}
def rc(seq):
    return "".join(complement.get(base, base) for base in reversed(seq))
```

```python
# search for boundary using bigwig and macs bed file
class JunInfor():
    def __init__(self):
        self.type=""
        self.chr=""
        self.loc="" #always left hand
        self.left_seq="" # All transform to plus strand
        self.right_seq="" # All transform to plus strand
    
    def toStr(self):
        return "{self.chr},{self.type},{self.loc},{self.left_seq},{self.right_seq}"

def getJunctionInfor(read):
    cigar_str = read.cigarstring
    chrom = read.reference_name
    start = read.reference_start
    end = read.reference_end
    query_seq = read.query_sequence
    m = re.match(r"^(\d+)[S](\d+)M$", cigar_str) # 5' (+, -)
    n = re.match(r"^(\d+)M(\d+)[S]$", cigar_str) # 3' (-, +)
    if m or n: 
        ji=JunInfor()
        ji.chr=chrom
        strand="+" 
        if read.is_reverse: strand = '-'
        if m and strand =="+":
            left_len = int(m.groups(0)[0])
            ji.loc=start-1
            ji.type=5
            ji.left_seq= query_seq[0:left_len]
            ji.right_seq= query_seq[left_len:]

        elif m and strand=="-":
            left_len = int(m.groups(0)[0])
            ji.loc=end
            ji.type=3
            ji.left_seq= rc(query_seq[left_len:])
            ji.right_seq= rc(query_seq[0:left_len])
        elif n and strand=="+":
            left_len = int(m.groups(0)[0])
            ji.loc=end
            ji.type=3
            ji.left_seq= query_seq[0:left_len]
            ji.right_seq= query_seq[left_len:]
        elif n and strand=="-":
            left_len = int(m.groups(0)[0])
            ji.loc=start-1
            ji.type=5
            ji.left_seq= rc(query_seq[left_len:]
            ji.right_seq= rc(query_seq[0:left_len:])
        else:
            return None

    else:
        return None

#       3|4    10|11
#  5'-ATG|AAT.AAG|GAC-3' 
#  3'-TAC|TTA.TTC|CTG-5'
read1=AlignedSegment()
read1.cigarstring="3S3M"
read1.reference_name = "chr1"
read1.reference_start = 4
read1.reference_end = 6
read1.query_sequence = "ATGAAT"

print(getJunctionInfor(read1).toStr())

read2=AlignedSegment()
read2.cigarstring="3M3S"
read2.reference_name = "chr1"
read2.reference_start = 8
read2.reference_end = 10
read2.query_sequence = "AAGGAC"

print(getJunctionInfor(read2).toStr())

read3=AlignedSegment()
read3.cigarstring="3S3M"
read3.reference_name = "chr1"
read3.reference_start = 8
read3.reference_end = 10
read3.query_sequence = "GTCCTT"

print(getJunctionInfor(read3).toStr())

read2=AlignedSegment()
read2.cigarstring="3M3S"
read2.reference_name = "chr1"
read2.reference_start = 4
read2.reference_end = 6
read2.query_sequence = "ATTCAT"

print(getJunctionInfor(read4).toStr())




    
```
