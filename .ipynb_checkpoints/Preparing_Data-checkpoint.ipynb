```python
# Download reference geneome 
# Genome sequences/annoation were download form here
# https://www.gencodegenes.org/human/
import os
def buildGenome(parent_dir, genome_name):
    REF_FA_URL="ftp://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_37/GRCh38.p13.genome.fa.gz"
    REF_GTF_URL="ftp://ftp.ebi.ac.uk/pub/databases/gencode/Gencode_human/release_37/gencode.v37.chr_patch_hapl_scaff.annotation.gtf.gz"
    if (os.path.isdir(parent_dir)):
        if (not os.path.isdir(parent_dir+"/"+genome_name)):
            os.system(f"mkdir {parent_dir}/{genome_name}")
        if (not os.path.isfile(parent_dir+"/"+genome_name+"/GRCh38.p13.genome.fa")):
            os.system(f"wget {REF_FA_URL} -P {parent_dir}/{genome_name}")
            os.system(f"unzip -d {parent_dir}/{genome_name}/GRCh38.p13.genome.fa.gz")
            os.system(f"samtools index {parent_dir}/{genome_name}/GRCh38.p13.genome.fa -@ 3")
            os.system(f"bwa index {parent_dir}/{genome_name}/GRCh38.p13.genome.fa")
        if (not os.path.isfile(parent_dir+"/"+genome_name+"/gencode.v37.chr_patch_hapl_scaff.annotation.gtf")):
            if (not os.path.isdir(parent_dir+"/"+genome_name+"/gft")):
                os.system(f"mkdir {parent_dir}/{genome_name}/gtf")
            if (not os.path.isfile(parent_dir+"/"+genome_name+"/gft/gencode.v37.chr_patch_hapl_scaff.annotation.gtf")):
                os.system(f"wget {REF_GTF_URL} -P {parent_dir}/{genome_name}/gtf")
                os.system(f"unzip -d {parent_dir}/{genome_name}/gtf/gencode.v37.chr_patch_hapl_scaff.annotation.gtf.gz")
        
buildGenome("/home/hqyone/mnt/2t/eccDNA/genome","hg38_p13")
```
